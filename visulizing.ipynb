```python
# replicate the above code for the duration column
du_df = yt_df.sort_values(by="duration", ascending=False, inplace=False)
du_df
```

```python
y_du = du_df['duration']
y_du_t = [datetime.strptime(y_du, "%H:%M:%S") for y_du in y_du] 
y_du_t = [(y_du_t.hour * 3600 + y_du_t.minute * 60 + y_du_t.second) for y_du_t in y_du_t]
y_du_t
```

```python
x_du = du_df['video_title'].str.split('- ')
x_du = [video[1] for video in x_du]
x_du
```

```python
fig, ax = plt.subplots(figsize=(17,15))

sn.barplot(x=y_du_t, y=x_du, width=0.5, palette="flare_r", ax=ax)
ax.set_xlabel('Duración')
ax.set_facecolor('white')
ax.bar_label(ax.containers[0], labels=y_du, color='black', fontsize=12, padding=1.5)
ax.set(xticks=[0, 3600, 5400, 7200, 9000, 10800, 12600, 14400])
ax.set(xticklabels=["", "01:00", "01:30", "02:00", "02:30", "03:00", "03:30", "04:00"])    

plt.grid(axis = 'x', color ='black', linestyle='--', linewidth = 0.5)
fig.tight_layout()
plt.savefig('youtube_vids_2nd_pull_seaborn_duration.jpg')
plt.show()
```

```python
dt_df = yt_df.sort_values(by="upload_date", ascending=False, inplace=False)
dt_df
```

```python
view_dt_df = dt_df[['upload_date','view_count']]
view_dt_df
```

```python
dt = dt_df['upload_date']
dt = [dt.split('-') for dt in dt]
dt = ['-'.join(dt[0:2]) for dt in dt]
dt
```

```python
view_dt_df.iloc[:,[0]] = dt
view_dt_df = view_dt_df.groupby('upload_date', as_index=False).sum()
view_dt_df
```

```python
#view_dt_df = view_dt_df.sort_values(by="upload_date", ascending=False, inplace=False)
#view_dt_df
```

```python
y_dt = view_dt_df['view_count']
y_dt
```

```python
x_dt = view_dt_df['upload_date']
x_dt
```

```python
fig, ax = plt.subplots(figsize=(12,8))

sn.barplot(x=x_dt, y=y_dt, width=0.3, palette='crest', ax=ax).set(title = 'Vistas por fecha de subida')
sn.lineplot(x=x_dt, y=y_dt, ax=ax, markers=True, dashes=False, color='red', linewidth=2.5, style="y_dt")

ax.set_ylabel('Vistas por millón')
ax.set_xlabel('Fecha de subida')
ax.set_facecolor('white')
ax.bar_label(ax.containers[0], labels=y_dt, color='black', fontsize=14, padding=4.5)


fig.tight_layout()

plt.grid(color ='black', linestyle='--', linewidth = 0.2)
plt.savefig('youtube_vids_2nd_pull_seaborn_upload_date.jpg')
plt.show()
```
