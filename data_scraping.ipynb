```python
def get_video_stats(video_id):

    url_video_stats = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_KEY
    response_video_stats = req.get(url_video_stats).json()

    view_count = response_video_stats['items'][0]['statistics']['viewCount']
    like_count = response_video_stats['items'][0]['statistics']['likeCount']
    comment_count = response_video_stats['items'][0]['statistics']['commentCount']
    

    url_video_content = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=contentDetails&key="+API_KEY
    response_video_content = req.get(url_video_content).json()

    duration = response_video_content['items'][0]['contentDetails']['duration']

    return view_count, like_count, comment_count, duration
```

```python
def get_videos(yt_df):
    
    pageToken = ""

    while 1:
        
        url = "https://www.googleapis.com/youtube/v3/search?key="+API_KEY+"&channelId="+CHANNEL_ID+"&part=snippet,id&order=date&maxResults=10000&"+pageToken

        response = req.get(url).json()
        time.sleep(5)

        for video in response['items']:

            if video['id']['kind'] == "youtube#video":

                video_id = video['id']['videoId']
                video_title = video['snippet']['title']
                video_title = str(video_title).replace("&amp;","")
                video_title = str(video_title).replace("é","e")
                video_title = str(video_title).replace("ó","o")
                video_title = str(video_title).replace("&quot","'")
                video_title = str(video_title).replace(";","")
                video_title = str(video_title).replace("&#39","'")
                upload_date = video['snippet']['publishedAt']
                upload_date = upload_date.split('T')[0]

                view_count, like_count, comment_count, duration = get_video_stats(video_id)

                if "El Método Rebord" in video['snippet']['title'] or "El Metodo Rebord" in video['snippet']['title'] or "Cónclave" in video['snippet']['title'] or "Conclave" in video['snippet']['title']:

                    time_string = duration
                    if "S" in time_string:
                        formater = datetime.strptime(time_string, 'PT%HH%MM%SS')
                        normal_time = formater.strftime('%H:%M:%S')
                        duration = normal_time
                    
                    elif "M" in time_string:
                        formater = datetime.strptime(time_string, 'PT%HH%MM')
                        normal_time = formater.strftime('%H:%M:%S')
                        duration = normal_time
                        
                    else:
                        formater = datetime.strptime(time_string, 'PT%HH')
                        normal_time = formater.strftime('%H:%M:%S')
                        duration = normal_time
                        

                    yt_df = yt_df.append({'video_id': video_id, 'video_title': video_title, 'upload_date': upload_date, 'view_count': view_count, 'like_count': like_count, 'comment_count': comment_count, 'duration': duration}, ignore_index=True)
        
        try:
            if response['nextPageToken'] != None: #if none, it means it reached the last page and break out of it
                pageToken = "pageToken=" + response['nextPageToken']
        
        except:
            break

    return yt_df
```

```python
yt_df = pd.DataFrame(columns=['video_id', 'video_title', 'upload_date', 'view_count', 'like_count', 'comment_count', 'duration'])
```

```python
yt_df = get_videos(yt_df)
```
